```python
import matplotlib.pyplot as plt

c_price = [39.44,39.91,41.54,40.07,39.53,38.62,37.89,37.27,38.05,37.91,37.40]
m_price = [39.8550,39.6700,39.9420,39.9580,39.8700,39.6580,39.4013,39.1421,39.0198,38.9079,38.7699]
xx = list(range(0,51))
plt.figure(facecolor='#24272d')
plt.rcParams["figure.figsize"] = (20,4)
plt.rcParams.update({"axes.grid" : True, "grid.color": "#24272d"})
ax = plt.axes()
ax.set_facecolor('#33363b')
ax.tick_params(color='black',grid_alpha=0.8)
plt.plot(c_price,color='#c1e0ff',  linestyle='dashed', label="Close Price")
plt.plot(m_price,color='#f78181',  linestyle='solid',  label="Mean Price")
plt.xticks(xx,color='darkgray')
plt.yticks(color='darkgray')
plt.show()
plt.savefig('d:/output_0010.png',dpi=200)
```
